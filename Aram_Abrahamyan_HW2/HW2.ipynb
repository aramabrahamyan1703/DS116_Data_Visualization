```python
sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize=(10, 5), facecolor='#f4f4f4')
plt.hist(pollution_df['PM2.5_AQI_Value'], bins=30, density=True, label='Histogram')
sns.kdeplot(pollution_df['PM2.5_AQI_Value'], color='darkred', fill=True, linewidth=2, label='Density Plot(KDE)')
plt.title('PM2.5 AQI Distribution with Density Overlay', weight='bold')
plt.xlabel('PM2.5 AQI Value', weight='bold')
plt.ylabel('Density', weight='bold')
txt="This plot represents the distribution of PM2.5 AQI values with density overlay."
plt.figtext(0.65, -0.1, txt, horizontalalignment='right')
plt.legend(["Histogram", "Density Plot (KDE)"])
plt.gca().set_facecolor('#eaeaf2');
#The colors in my IDE are showing in black by default, in your case it should work fine))))(have checked with different IDE)
```
